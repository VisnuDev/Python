```python
#1.	Write a Python Program to Add Two Matrices?


rows= int(input("Enter the number of rows: "))
cols= int(input("Enter the number of columns: "))


print("Enter the elements of Matrix A: ")

matrix_a= [[int(input())  for c in range(cols) ]  for r in range(rows)]

print("Matrix A:")
for i in matrix_a:
    print(i)
    

print("Enter the elements of Matrix B: ")

matrix_b= [[int(input())  for c in range(cols) ]  for r in range(rows)]

print("Matrix B: ")
for i in matrix_b:
    print(i)
    
    
    
add_matrix =[[0 for c in range(cols)]  for r in range(rows)]


for i in range(rows):
    for j in range(cols):
        add_matrix[i][j] = matrix_a[i][j] + matrix_b[i][j]
        
        

print("The resultant matrix after addition: ")

for x in add_matrix:
    print(x)
```

```python
#3.	Write a Python Program to Transpose a Matrix?



def matTranspose(mat):
    rows= len(mat)
    cols= len(mat[0])
    trans_mat= [[0 for c in range(rows)]  for r in range(cols)]
    for i in range(rows):
        for j in range(cols):
            trans_mat[j][i]=mat[i][j]
    return trans_mat
            


rows= int(input("Enter the number of rows: "))
cols= int(input("Enter the number of columns: "))

print("Enter the elements of the matrix: ")

mat= [[int(input()) for c in range(cols)]  for r in range(rows)]


print("Before Transpose: ")
for x in mat:
    print(x)
    

    
trans_mat= matTranspose(mat)

print("After Transpose: ")
for x in trans_mat:
    print(x)
```

```python
#4.	Write a Python Program to Sort Words in Alphabetic Order?


def sortWords(s):
    s= s.lower()
    words= s.split()
    words.sort()
    
    print("After sorting the words alphabetically: ")
    
    for w in words:
        print(w, end=" ")
        
    
    
    
string= input("Enter the string: ")


sortWords(string)

```

```python
#5.	Write a Python Program to Remove Punctuation From a String?

def removePunct(s):
    new_string=""
    for c in s:
        if c.isalnum() or c.isspace():
            new_string = new_string + c
        else:
            pass
    return new_string
        
        
string = input("Enter the string: ")

no_punt_str= removePunct(string)

print("After removing puntuations: ")

print(no_punt_str)
```
